Task-4: Analyze and visualize sentiment patterns in social media data to understand public opinion and attitudes towards specific topics or brands.

```python
import pandas as pd

data = pd.read_csv('twitter_training.csv', header=None, names = ['ID', 'Topic', 'Sentiment', 'Text'])
display(data.head())
print(data.shape)
print(data.info()) 
print(data.isnull().sum())
```

```python
import matplotlib.pyplot as plt

sentiment_counts = data['Sentiment'].value_counts()
plt.figure(figsize=(8, 6), dpi = 200)
sentiment_counts.plot(kind='bar', color=['lime', 'orangered', 'magenta', 'cyan']) 
plt.title('\nOverall Sentiment Distribution\n', fontweight = 'bold', fontsize = 23)
plt.xlabel('\nSentiment\n', fontweight = 'bold', fontsize = 12)
plt.ylabel('Number of Tweets', rotation = 0, labelpad = 50, fontweight = 'bold', fontsize = 10)
plt.xticks(rotation=0)
plt.show() 
```

```python
plt.figure(figsize=(8, 8), dpi = 200)
sentiment_counts.plot(kind='pie', autopct='%1.1f%%', colors=['lime', 'yellow', 'magenta', 'cyan'], startangle = 60, shadow = True)
plt.title('\nOverall Sentiment Distribution', fontweight = 'bold', fontsize = 25)
plt.ylabel('')
plt.show() 
```

```python
# Display unique topics and their counts
topic_counts = data['Topic'].value_counts()
topic_counts
```

```python
import matplotlib.pyplot as plt
from wordcloud import WordCloud

def generate_wordcloud(text, title):
    wordcloud = WordCloud(width=800, height=400, background_color='white').generate(text)
    plt.figure(figsize=(10, 5), dpi = 200)
    plt.imshow(wordcloud, interpolation='bilinear')
    plt.title(f"\n{title}\n", fontsize = 26, weight='bold')
    plt.axis('off')
    plt.show()

data = data.dropna(subset=['Text'])

positive_text = ' '.join(data[data['Sentiment'] == 'Positive']['Text'])
generate_wordcloud(positive_text, 'Word Cloud for Positive Tweets')

negative_text = ' '.join(data[data['Sentiment'] == 'Negative']['Text'])
generate_wordcloud(negative_text, 'Word Cloud for Negative Tweets')

neutral_text = ' '.join(data[data['Sentiment'] == 'Neutral']['Text'])
generate_wordcloud(neutral_text, 'Word Cloud for Neutral Tweets')
```
